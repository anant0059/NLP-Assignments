```python
df=df[['Title']]
df.head()
```

```python
sent_list=df.values.tolist()
sent_list[0]

```

```python
def str_list(Input_Sentence): 
    output_ = Input_Sentence.split()
    return output_
```

```python
sent_word_list=[]
for i in sent_list:
    sent_word_list.append(str_list(i[0]));
```

```python
sent_word_list
```

```python
dic={}
```

```python
def av_size(sentence):
    size=0
    for i in sentence:
        size=size+len(i)
        for j in i:
            dic[j]=1
    return size/len(sentence)
```

```python
avarage_size=av_size(sent_word_list)
print(avarage_size)
```
